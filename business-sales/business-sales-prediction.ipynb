# Buisness Sales Prediction using Linear Regression

This project predicts business sales based on advertising, store visitores and competitor price.


```python
import pandas as pd
import numpy as np

# Random seed for reproducibility
np.random.seed(42)

# Generate dataset
data = {
    "Month": np.arange(1, 13),
    "Advertising": np.random.randint(5000, 20000, 12),
    "Store_Visitors": np.random.randint(200, 800, 12),
    "Competitor_Price": np.random.randint(50, 150, 12),
}

# Simulate Sales (target) as a combination of features + some noise
data["Sales"] = (0.5 * data["Advertising"] +
                 10 * data["Store_Visitors"] -
                 20 * data["Competitor_Price"] +
                 np.random.randint(-2000, 2000, 12))

# Create DataFrame
df = pd.DataFrame(data)

# Display
df
```

```python
from sklearn.linear_model import LinearRegression
```

```python
# 1. Create the model
model = LinearRegression()

# 2. Train the model using all the data
X = df[["Advertising", "Store_Visitors", "Competitor_Price"]]  # Features
y = df["Sales"]  # Target

model.fit(X, y)
```

```python
# Predict Sales for the same data
predictions = model.predict(X)

# See the predictions
print(predictions)
```

```python
from sklearn.metrics import mean_absolute_error

# Calculate Mean Absolute Error
mae = mean_absolute_error(y, predictions)
print("Mean Absolute Error:", mae)
```

```python
df_compares = df.copy()
df_compares["Predicted_Sales"] = predictions
df_compares[["Sales", "Predicted_Sales"]]
```
