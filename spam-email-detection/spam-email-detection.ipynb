# Spam Email Detection using Machine Learning

This project predicts whether a message is spam or not spam using text data.


```python
import pandas as pd

data = {
    'Message': [
        'Win a free iPhone now',
        'Congratulations you won a prize',
        'Call me when you are free',
        'Let us meet tomorrow',
        'Limited offer buy now',
        'Hey are you coming to class',
        'You have won cash reward',
        'Are we meeting today'
    ],
    'Spam': [1, 1, 0, 0, 1, 0, 1, 0]
}

spam = pd.DataFrame(data)
spam
```

```python
from sklearn.feature_extraction.text import CountVectorizer    #count vectorixer will convert text to forst list then number as ml cannot words

vectorizer = CountVectorizer()
X = vectorizer.fit_transform(spam['Message'])
y = spam['Spam']
```

```python
from sklearn.naive_bayes import MultinomialNB

# Create the model
model = MultinomialNB()

# Train the model
model.fit(X, y)

print("Model trained successfully")
```

```python
predictions = model.predict(X)
predictions
```

```python
from sklearn.metrics import accuracy_score

accuracy = accuracy_score(y, predictions)
print("Model Accuracy:", accuracy)
```
