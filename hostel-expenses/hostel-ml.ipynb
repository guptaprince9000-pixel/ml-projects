```python
daily = hostel.groupby('Date')['Amount'].sum().reset_index()
daily
```

```python
daily['Day'] = range(1, len(daily)+1)
daily
```

```python
X = daily[['Day']]        # feature
y = daily['Amount']       # target
```

```python
from sklearn.linear_model import LinearRegression
```

```python

# 1. Create the model
model = LinearRegression()

# 2. Train the model
model.fit(X, y)

# 3. Make predictions for the same days
predictions = model.predict(X)

# 4. See the predictions
print(predictions)
```
