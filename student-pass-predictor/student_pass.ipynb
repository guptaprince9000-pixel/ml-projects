Project: Student Pass Prediction
Goal: Predict whether a student will pass or fail based on study hours

## Dataset Description

This dataset contains:
- Study Hours: Number of hours a student studies per day
- Pass: Whether the student passed the exam (1 = Pass, 0 = Fail)


| Study Hours | Pass |
|------------|------|
| 1 | 0 |
| 2 | 0 |
| 3 | 0 |
| 4 | 1 |
| 5 | 1 |
| 6 | 1 |
| 7 | 1 |


```python
import pandas as pd

df=pd.DataFrame({'StudyHours':[1,2,3,4,5,6,7], 'Pass':[0,0,0,1,1,1,1]})
```

```python
df
```

```python
# Feature (input)
X = df[["StudyHours"]]

# Target (output)
y = df["Pass"]

X, y
```

```python
df["StudyHours"].shape
```

```python
from sklearn.linear_model import LogisticRegression
## this LogisticRegression is beign used fro given yes/no or pass/fail
```

```python
model = LogisticRegression()
model.fit(X, y)

##model = an empty brain
# fit = teaching the brain using your data
# The model learned:
# When hours are low → fail
# When hours are high → pass
```

```python
# Predict for a student who studies 5 hours
prediction = model.predict(X)
prediction
```

```python
for hours, result in zip(X["StudyHours"], prediction):
    if result == 1:
        print(f"Studied {hours} hours → PASS")
    else:
        print(f"Studied {hours} hours → FAIL")
```

```python
from sklearn.metrics import accuracy_score

accuracy = accuracy_score(y, prediction)
print("Model Accuracy:", accuracy)
```
