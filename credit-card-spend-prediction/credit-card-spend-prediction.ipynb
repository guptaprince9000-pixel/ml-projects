# Credit Card Spend Prediction using Linear Regression

This project predicts monthly credit card spending based on age and income.


```python
import pandas as pd

data = {
    'Age': [22, 25, 28, 30, 35, 40, 45, 50],
    'Income': [20000, 25000, 30000, 40000, 50000, 60000, 70000, 80000],
    'MonthlySpend': [5000, 7000, 9000, 12000, 15000, 18000, 20000, 23000]
}

credit = pd.DataFrame(data)
credit
```

```python
X=credit[['Age','Income']]
Y=credit['MonthlySpend']
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()  # 1. Create the model
model.fit(X, Y)             # 2. Train the model
```

```python
predictions=model.predict(X)
predictions
```

```python
from sklearn.metrics import mean_absolute_error

error = mean_absolute_error(Y, predictions)
print("Mean Absolute Error:", error)
```
