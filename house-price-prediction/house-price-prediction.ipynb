# House Price Prediction using Linear Regression

This project predicts house prices based on area and number of bedrooms.


```python
import pandas as pd

data = ({'Area': [800, 1000, 1200, 1500, 1800, 2000, 2200, 2500],'Bedrooms': [1, 2, 2, 3, 3, 3, 4, 4],'Price': [40, 50, 60, 75, 90, 100, 115, 130]})

house = pd.DataFrame(data)
house
```

```python
X = house[['Area', 'Bedrooms']]   # features
y = house['Price']                # target
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(X, y)
```

```python
predictions = model.predict(X)
predictions
```

```python
from sklearn.metrics import mean_absolute_error

error = mean_absolute_error(y, predictions)
error
```
