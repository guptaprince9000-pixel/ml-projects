# Loan Approval Prediction using Logistic Regression

This project predicts whether a loan will be approved based on applicant information.


```python
import pandas as pd

data = {
    'ApplicantIncome': [2500, 3000, 4000, 5000, 6000, 3500, 4500, 7000],
    'CreditHistory': [1, 0, 1, 1, 0, 1, 0, 1],
    'LoanAmount': [100, 120, 150, 200, 250, 130, 170, 300],
    'LoanApproved': [1, 0, 1, 1, 0, 1, 0, 1]
}

loan = pd.DataFrame(data)
loan
```

```python
X = loan[['ApplicantIncome', 'CreditHistory', 'LoanAmount']]  # features
y = loan['LoanApproved']                                      # target
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression()
model.fit(X, y)
```

```python
predictions = model.predict(X)
predictions
```

```python
from sklearn.metrics import accuracy_score

accuracy = accuracy_score(y, predictions)
print("Model Accuracy:", accuracy)
```
